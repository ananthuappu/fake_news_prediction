Removal of stopwords and stemming

```python
title=X['title']
```

```python
title
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
stop_words = set(stopwords.words('english'))
```

```python
def text_preprocessing(data_df):
    title_corpus=[]
    for i in range(0, len(data_df)):
    #     print(title[i])
    #     break
    #     Remove all characters other than letters.
        first_cut_title=re.sub('[^a-zA-Z]', ' ',data_df[i])
    #     break
        word_tokens = word_tokenize(first_cut_title)
        filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]
        filtered_sentence = []
        for w in word_tokens:
            if w not in stop_words:
                filtered_sentence.append(w)
        filtered_sentence=' '.join(filtered_sentence)
        title_corpus.append(filtered_sentence)
    return title_corpus
#     break

# print(word_tokens)
# print(filtered_sentence)
# print(first_cut_title)
```

```python
title_corpus=text_preprocessing(title)
```

```python
title_corpus[5]
```

Data Conversion (We are dealing with text data so we shld convert it into numerical form)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=5000,ngram_range=(1, 3))
X = vectorizer.fit_transform(title_corpus) # Resassigning cleaned feature as X
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DT_classifier=DecisionTreeClassifier()
```

```python
DT_classifier.fit(X_train, y_train)
```

```python
from sklearn import metrics
pred = DT_classifier.predict(X_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)
```
